```python
with open('med_news_p3.txt', 'r') as myfile:
  data = myfile.read()

soup = BeautifulSoup(data, 'html.parser')

print(data)

#scraped 
```

```python
for med_articles in soup.find_all("li", {"class": "written"}):
    print(med_articles)
    break
#path one for the website 2005-------------------------------------------------- step 2 
```

```python
m_articles=[]

for med_articles in soup.find_all("li", {"class": "written"}):
    article={}
    article['title']=med_articles.find("a").text
    article['class']=med_articles.find("span",{"class":"story_hits"}).text
    m_articles.append(article)
    
print(m_articles)
########path 1 
#articles in the website are uunder two different types of titles because of the way the links were put together.
```

```python
with open ('csvfile.csv','w') as art_file: 
    writer = csv.writer(art_file)
    for i in m_articles:
        line=(i["title"],i["class"])
        writer.writerow(line)
    #creating a csv by making a line with commas ----------------------------------------------------------------------
```
